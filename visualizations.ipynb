```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the spreadsheet (replace with your file path)
data = pd.read_csv("copppy - Sheet13.csv")
data = data.drop(columns=["Average accuracy","no. of elections participated"])

# Extract election names (assuming they're in the first row)
elections = list(data)[1:]  # Skip the first "organization" column

# Create the plot
plt.figure(figsize=(12, 6))

for i, organization in data.iterrows():
    accuracy_values = organization[1:].tolist()  # Get accuracy values for each election
    plt.plot(elections, accuracy_values, label=organization[0])  # Plot line for each organization

# Customize the plot
plt.xlabel("Election")
plt.ylabel("Accuracy (%)")
plt.title("Performance Curves of Organizations in 10 Elections")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
plt.legend()
plt.grid(True)

# Show or save the plot
plt.show()  # To display the plot interactively
# plt.savefig("performance_curves.png")  # To save as an image
```
